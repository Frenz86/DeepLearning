Wildcards match repeated characters
"\w\w\w". Because this is such a common need, there is a specific syntax to match repetitions

```python
#regex = re.compile(r'\w\w\w')
regex = re.compile(r'\w{3}')
regex.findall('The quick brown fox')
```

```python
#"+" character will match one or more repetitions
```

```python
regex = re.compile(r'\w+')
regex.findall('The quick brown fox')
```

Character	Description	Example
?	Match zero or one repetitions of preceding	"ab?" matches "a" or "ab"
*	Match zero or more repetitions of preceding	"ab*" matches "a", "ab", "abb", "abbb"...
+	Match one or more repetitions of preceding	"ab+" matches "ab", "abb", "abbb"... but not "a"
{n}	Match n repetitions of preeeding	"ab{2}" matches "abb"
{m,n}	Match between m and n repetitions of preceding	"ab{2,3}" matches "abb" or "abbb"

We have changed "\w+" to "[\w.]+", so we will match any alphanumeric character or a period. With this more flexible expression, we can match a wider range of email addresses (though still not all – can you identify other shortcomings of this expression?).

```python
text = """ Questo corso BIGDATA LAB spacca di brutto!
        posso chiamare Daniele ai numeri +39 333-333-1234
        oppure al +39 333-333-1235
        """


pattern = r'(\d{2}) (\d{3})-(\d{3}-\d{4})'

match_ = re.findall(pattern, text)

print(match_)
```

```python
##range
text = """ Questo corso FAV spacca di brutto!
        posso chiamare Daniele ai numeri +39 333-333-1234
        oppure al +39 333-33333-12356
        """


pattern = r'(\d{2}) (\d{3})-(\d{3,5}-\d{3,5})'

match_ = re.findall(pattern, text)

print(match_)
```

Riferimenti utili per usare la sintassi per creare regular expression

| Character | Description                 || Character | Description                     |
|-----------|-----------------------------||-----------|---------------------------------|
| ``"\d"``  | Match any digit             || ``"\D"``  | Match any non-digit             |
| ``"\s"``  | Match any whitespace        || ``"\S"``  | Match any non-whitespace        |
| ``"\w"``  | Match any alphanumeric char || ``"\W"``  | Match any non-alphanumeric char |

This is *not* a comprehensive list or description; for more details, see Python's [regular expression syntax documentation](https://docs.python.org/3/library/re.html#re-syntax).

### Possibile uso in NLP delle Regular Expression

* Eliminare tutto ciò che non è alfanumerico e spazio con ^\w\s

* Sostituisci gli a capo "\n"," "

* Eliminare le maiuscole con metedo .lower



```python
import re
text = ("""Ciao, come STAI????!!...,
scrivo test a caso%%&&$££, 1233£""")
```

```python
text = re.sub(r'[^\w\s]','',text) #^negazione \w=any alphanumeric char \s=any whitespace
##                                 ----------> tutto ciò che non è alfanumerico
text = text.replace("\n","")
text = text.lower()
text
```
