> ## RNN

```python
# import
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Flatten, TimeDistributed, LSTM, Dense
```

```python
# build model
model = Sequential()
model.add(SimpleRNN(128, input_shape=(5, 10)))
model.add(Dense(1))

model.summary()
```

```python
# print all matrices
for x in  model.layers[0].weights:
    print(x.name,'-->', x.shape)
```

## Deep Vertical RNNs

```python
# build model
model = Sequential()
model.add(SimpleRNN(128, input_shape=(5, 10), return_sequences=True))
model.add(SimpleRNN(128, return_sequences=True))
model.add(SimpleRNN(128, return_sequences=False))
model.add(Dense(1))

model.summary()
```

```python
# print all matrices
for x in  model.layers[0].weights:
    print(x.name,'-->', x.shape)
```

## LSTM

```python
# build model
model = Sequential()
model.add(LSTM(128, input_shape=(5, 10)))
model.add(Dense(1))

model.summary()
```

```python
# print all matrices
for x in  model.layers[0].weights:
    print(x.name,'-->', x.shape)
```
